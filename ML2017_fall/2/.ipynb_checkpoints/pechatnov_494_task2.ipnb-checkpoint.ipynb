```python
lam = 0.1

ts['val_boxcox'] = (np.array(ts['val']) ** lam + 1) ** (1 / lam) 
ts['val_boxcox'].plot(figsize=(15,5))
plt.show()

kpss_stat = stattools.kpss(np.array(ts['val_boxcox']))
print('diff 1 The KPSS test statistic %f, \npvalue: %.2f \nmaximal Lag: %i \ntable distribution: %s' % 
      (kpss_stat[0], kpss_stat[1], kpss_stat[2], kpss_stat[3]))
df_stat = stattools.adfuller(np.array(ts['val_boxcox']))
print('The Augmented Dickey-Fuller test statistic %f, \npvalue: %.2f \nmaximal Lag: %i \ncritical values: %s' % 
      (df_stat[0], df_stat[1], df_stat[2], df_stat[4]))
```

Возьмем p и q равными 1, 2 или 27

upd 27 работает бесконечно долго, отбрасываем

```python
def arima_fit(*pdq):
    return sm.tsa.ARIMA(train_ts, (*pdq)).fit()

def arima_fit_aic(*pdq):
    try:
        aic = arima_fit(pdq).aic
        print("ARIMA", pdq, "->", aic)
        return aic
    except:
        return 1e100
```

```python
from itertools import product
pdq_cands = product([1, 2], [0], [1, 2])

aics = [(arima_fit_aic(*par), par) for par in pdq_cands]
```

```python
best_pdq = min(aics)[1]
min(aics)
```

```python
model = arima_fit(best_pdq)
model.resid.plot(figsize=(10,4))
plt.show()

df_stat = stattools.adfuller(np.array(model.resid))
print('The Augmented Dickey-Fuller test statistic %f, \npvalue: %.2f \nmaximal Lag: %i \ncritical values: %s' % 
      (df_stat[0], df_stat[1], df_stat[2], df_stat[4]))

stat = stats.wilcoxon(model.resid.values)
print('WILCOXON Statistic: %f\npvalue: %.2f' %(stat[0], stat[1]))
```

Остатки не стационарны =( Но с несмещенностью все небезнадежно

**II. Find the best parameters of ARIMA automaticaly** 
* 0) read data for TS under consideration and use the first 80% of all length of the TS as train TS.  
* 1) plot the graph of the train TS, preprocess data (delete tail, make calendar transformation if needed)
* 2) stabilize dispersion of train TS through box-cox transformation if needed.
* 3) Run grid search for ARIMA parameters (use transformed train TS) according to AIC. Use at least 10 distinct values for $p,q,d$. How many time does it take?
* 4) Build forecast for the initial TS with best ARIMA model selected automaticaly at step 2 (you need to transform it like in step 2).
* 5) Build forecast interval for the last 20% of transformed TS. Calculate loss MACAPE of your forecast. Does the forecast look appropriate?

```python
def build_best_arima_model(pdq_cands=None):
    if pdq_cands is None:
        pdq_cands = product(range(10), range(5), range(10))
    aics = [(arima_fit_aic(*par), par) for par in pdq_cands]
    pdq = min(aics)[1]
    model = arima_fit(*pdq)
    return model
                              
def build_and_show_arima_forecast(pdq_cands=None):
    model = build_best_arima_model(pdq_cands)
    test_ts["val_forecast"], test_ts["val_err"] = model.forecast(steps=len(test_ts))[:2]
    ts["val_forecast"] = test_ts['val_forecast']
    ts["val_forecast_err"] = test_ts['val_err']
    plot_forecast(ts)
    
    
build_and_show_arima_forecast()
```
